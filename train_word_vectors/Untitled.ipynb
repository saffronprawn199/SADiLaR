```python
from gensim.test.utils import common_texts
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

model =  FT_gensim.load("/Users/rhynostrydom/PycharmProjects/SADiLaR/train_word_vectors/fastText_models/fastText_train_epochs_trained_10_afrikaans_training_data_SG_embedding_size_300")

# labels = [i for i in model.wv.vocab.keys()]
labels =list(model.wv.index_to_key)
tokens = vectors

tsne_model = TSNE(init='pca',learning_rate='auto')
new_values = tsne_model.fit_transform(tokens)
```

```python
STOP_WORDS = set(
    """
'n
aan
af
al
as
baie
by
daar
dag
dat
die
dit
een
ek
en
gaan
gesê
haar
het
hom
hulle
hy
in
is
jou
jy
kan
kom
ma
maar
met
my
na
nie
om
ons
op
saam
sal
se
sien
so
sy
te
toe
uit
van
vir
was
wat
ŉ
""".split()
)
```

```python
# apply stop words check
labels_new= [word for word in labels if word not in STOP_WORDS]
```

```python
x = []
y = []
for value in new_values:
    x.append(value[0])
    y.append(value[1])


plt.figure(figsize=(7, 5)) 
for i in range(new_values.shape[0]):
    plt.scatter(x[i],y[i])
    plt.annotate(labels[i],
                 xy=(x[i], y[i]),
                 xytext=(5, 2),
                 textcoords='offset points',
                 ha='right',
                 va='bottom')
```
